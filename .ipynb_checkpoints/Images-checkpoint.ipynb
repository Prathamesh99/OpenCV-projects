```python
img1 = cv2.imread('DATA/dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2, (600,600))
```

```python
x_offset = 934 - 600
y_offset = 1401 - 600
```

```python
rows, cols, channels = img2.shape
```

```python
roi = img1[y_offset:1401,x_offset:943]
```

```python
plt.imshow(roi)
```

```python
img2gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img2gray, cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2gray)
```

```python
plt.imshow(mask_inv, cmap='gray')
```

```python
white_back = np.full(img2.shape, 255, dtype=np.uint8)
```

```python
bk = cv2.bitwise_or(white_back, white_back, mask=mask_inv)
```

```python
bk.shape
```

```python
plt.imshow(bk)
```

```python
fg = cv2.bitwise_or(img2, img2, mask=mask_inv)
```

```python
plt.imshow(fg)
```

```python
final_roi = cv2.bitwise_or(roi, fg)
```

```python
plt.imshow(final_roi)
```

```python
large_img = img1
small_img = final_roi
```

```python
large_img[y_offset:y_offset+small_img.shape[0], x_offset:x_offset+small_img.shape[1]] = small_img
```

```python
plt.imshow(large_img)
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img, 200,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img, 255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11,8)
show_pic(th2)
```

```python
blended = cv2.addWeighted(th1, 0.6, th2,0.4,0)
show_pic(blended)
```

```python
def load_img():
    img = cv2.imread('DATA/bricks.jpg').astype(np.float32)/255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
i = load_img()
display_img(i)
```
